Read from file

```python
import os
```

```python
os.makedirs("data", exist_ok=True)
```

```python
datafile = os.path.join("data", "house_tiny.csv")
with open(datafile, "w") as f:
  f.write('''NumRooms, RoofType, Price
  NA,NA, 1275002
  2, NA, 10600
  4, slate, 178100
  NA,NA, 14000''')
```

```python
import pandas as pd

df = pd.read_csv(datafile)
df
```

```python
inputs,targets = df.iloc[:, 0:2], df.iloc[:, -1]
```

```python
inputs
```

```python
targets
```

Data Preprocessing

```python
roofType = inputs.columns[-1]
inputs = pd.get_dummies(inputs, columns=[roofType], dummy_na=True)
inputs
```

```python
numRooms = inputs.columns[0]
inputs[numRooms] = pd.to_numeric(inputs[numRooms], errors="coerce")
inputs[numRooms] = inputs[numRooms].fillna(inputs[numRooms].mean())
inputs
```
